```python
import gradio as gr
import requests
import os

# --- 1. API é‡‘é‘°è¨­å®šèˆ‡ç’°å¢ƒè®Šæ•¸è™•ç† ---
# å»ºè­°å°‡é‡‘é‘°è¨­å®šç‚ºç’°å¢ƒè®Šæ•¸ 'AUDDIO_API_KEY'
# å¦‚æœç’°å¢ƒè®Šæ•¸æœªè¨­ç½®ï¼Œå°‡ä½¿ç”¨ç¨‹å¼ç¢¼ä¸­ç¡¬ç·¨ç¢¼çš„å‚™ç”¨é‡‘é‘°ã€‚
API_KEY = os.environ.get("AUDDIO_API_KEY", "3519b325d66d158312f4677cc81f5d2e")

def recognize_song(file_path: str):
    """
    æ¥æ”¶æª”æ¡ˆè·¯å¾‘ (string)ï¼Œé–‹å•Ÿä¸¦è®€å–æª”æ¡ˆå…§å®¹ï¼Œä½¿ç”¨ Audd.io API é€²è¡ŒéŸ³æ¨‚è¾¨è­˜ã€‚
    æ­¤å‡½æ•¸åŒ…å«è©³ç´°çš„éŒ¯èª¤é™¤éŒ¯è¼¸å‡ºã€‚
    """
    url = "https://api.audd.io/"

    if API_KEY is None or API_KEY == "":
        return "éŒ¯èª¤ï¼šAudd.io API é‡‘é‘°æœªè¨­ç½®ã€‚è«‹æª¢æŸ¥ç’°å¢ƒè®Šæ•¸æˆ–ç¨‹å¼ç¢¼ã€‚"

    # æª¢æŸ¥ Gradio å‚³å…¥çš„æ˜¯å¦ç‚ºå­—ä¸²è·¯å¾‘ (è§£æ±º 'NamedString' éŒ¯èª¤)
    if not isinstance(file_path, str):
        return "è¼¸å…¥éŒ¯èª¤ï¼šé æœŸæ¥æ”¶æª”æ¡ˆè·¯å¾‘å­—ä¸²ï¼Œä½†æ¥æ”¶åˆ°å…¶ä»–é¡å‹ã€‚"

    # æ ¸å¿ƒè™•ç†ï¼šä»¥äºŒé€²åˆ¶æ¨¡å¼è®€å–æª”æ¡ˆä¸¦ç™¼é€è«‹æ±‚
    try:
        with open(file_path, 'rb') as f:
            files = {
                # å‚³éæª”æ¡ˆå…§å®¹çµ¦ API
                'file': (os.path.basename(file_path), f.read(), 'audio/mpeg')
            }
            data = {
                'api_token': API_KEY,
                'return': 'spotify,apple_music,youtube,deezer'
            }

            response = requests.post(url, data=data, files=files)
            # æª¢æŸ¥ HTTP ç‹€æ…‹ç¢¼ã€‚å¦‚æœä¸æ˜¯ 200ï¼Œç«‹åˆ»æ‹‹å‡ºéŒ¯èª¤ã€‚
            response.raise_for_status()
            result = response.json()

    except FileNotFoundError:
        return f"### âŒ æª”æ¡ˆéŒ¯èª¤\néŒ¯èª¤ï¼šæ‰¾ä¸åˆ°æª”æ¡ˆè·¯å¾‘ {file_path}ã€‚"
    except requests.exceptions.HTTPError as e:
        return f"### âŒ HTTP é€£ç·šéŒ¯èª¤\nAPI ä¼ºæœå™¨å›å‚³ç‹€æ…‹ç¢¼ï¼š{e.response.status_code}ã€‚\n\n**ğŸ’¡ æ’æŸ¥å»ºè­°ï¼š** è‹¥æ˜¯ 401/403ï¼Œè«‹æª¢æŸ¥æ‚¨çš„ API é‡‘é‘°æ˜¯å¦æœ‰æ•ˆã€‚"
    except requests.exceptions.RequestException as e:
        return f"### âŒ ç¶²è·¯é€£ç·šéŒ¯èª¤\nç„¡æ³•é€£æ¥åˆ° Audd.io ä¼ºæœå™¨ï¼š{e}"
    except Exception as e:
        return f"### âŒ æœªçŸ¥éŒ¯èª¤\næª”æ¡ˆè®€å–æˆ– JSON è§£æéŒ¯èª¤ï¼š{e}"

    # --- 2. çµæœè§£æèˆ‡è©³ç´°é™¤éŒ¯ ---

    # 1. æª¢æŸ¥ Audd.io å…§éƒ¨ç‹€æ…‹æ˜¯å¦ç‚º 'error'
    if result.get("status") == "error":
        error_info = result.get("error", {})
        error_message = error_info.get("error_message", "Audd.io å›å‚³äº†æœªçŸ¥çš„éŒ¯èª¤è¨Šæ¯ã€‚")
        error_code = error_info.get("error_code", 'N/A')

        output = f"### âŒ æœå‹™éŒ¯èª¤ (Audd.io å…§éƒ¨)\n"
        output += f"**éŒ¯èª¤ä»£ç¢¼ï¼š** {error_code}\n"
        output += f"**éŒ¯èª¤è¨Šæ¯ï¼š** {error_message}\n"
        output += "\n--- \n**ğŸ’¡ æ’æŸ¥å»ºè­°ï¼š** è«‹ä»”ç´°æª¢æŸ¥æ‚¨çš„ API é‡‘é‘°æ˜¯å¦æ­£ç¢ºã€æ˜¯å¦éæœŸï¼Œæˆ–ç”¨é‡æ˜¯å¦å·²æ»¿ã€‚"
        return output

    # 2. æª¢æŸ¥æ˜¯å¦æˆåŠŸè¾¨è­˜åˆ°çµæœ (status="success" ä½† result ç‚º null)
    if not result.get("result"):
        return "### â“ ç„¡æ³•è¾¨è­˜\næˆåŠŸé€£ç·šè‡³ Audd.ioï¼Œä½†ç„¡æ³•åœ¨è³‡æ–™åº«ä¸­æ‰¾åˆ°åŒ¹é…çš„éŸ³æ¨‚ã€‚\n\n**ğŸ’¡ æ’æŸ¥å»ºè­°ï¼š** è«‹å˜—è©¦æ›´æ¸…æ™°ã€é•·åº¦è¶³å¤ ï¼ˆè‡³å°‘ 10-15 ç§’ï¼‰çš„éŸ³æ¨‚æª”æ¡ˆï¼Œæˆ–ç¢ºèªéŸ³é »ä¸­æ˜¯å¦åŒ…å«å¯è¾¨è­˜çš„éŸ³æ¨‚ã€‚"

    # --- 3. æˆåŠŸè¾¨è­˜çµæœæ ¼å¼åŒ– ---
    r = result["result"]

    # ä½¿ç”¨ Markdown æ ¼å¼åŒ–è¼¸å‡º
    output = f"### âœ… ğŸ¶ éŸ³æ¨‚è¾¨è­˜çµæœ\n"
    output += f"**æ­Œæ›²ï¼š** {r.get('title', 'N/A')}\n"
    output += f"**æ­Œæ‰‹ï¼š** {r.get('artist', 'N/A')}\n"
    output += f"**å°ˆè¼¯ï¼š** {r.get('album', 'N/A')}\n"

    output += "\n---\n"
    output += "**ç›¸é—œé€£çµï¼š**\n"

    # æ·»åŠ å¤–éƒ¨é€£çµ
    if r.get("spotify"):
        output += f"* [ğŸ§ Spotify é€£çµ]({r['spotify']['external_urls']['spotify']})\n"
    if r.get("youtube"):
        output += f"* [â–¶ï¸ YouTube é€£çµ]({r['youtube']['url']})\n"
    if r.get("deezer"):
        output += f"* [ğŸµ Deezer é€£çµ]({r['deezer'].get('link', '#')})\n"
    if r.get("apple_music"):
        output += f"* [ğŸ Apple Music é€£çµ]({r['apple_music']['url']})\n"

    return output

# --- 4. Gradio UI è¨­ç½® ---
with gr.Blocks(title="ç·šä¸ŠéŸ³æ¨‚è¾¨è­˜å™¨ Audd.io ç‰ˆ") as demo:
    gr.Markdown("# ğŸµ ç·šä¸ŠéŸ³æ¨‚è¾¨è­˜å™¨ Audd.io ç‰ˆ\nä¸Šå‚³ MP3 / MP4 / WAV æª”æ¡ˆå³å¯è¾¨è­˜æ­Œæ›²")

    # gr.File é è¨­å‚³å›æª”æ¡ˆè·¯å¾‘å­—ä¸² (file_path)ï¼Œé€™æ˜¯ä¿®æ­£ 'NamedString' éŒ¯èª¤çš„é—œéµ
    audio_input = gr.File(label="ä¸Šå‚³éŸ³æ¨‚æª”æ¡ˆ (æ”¯æ´ MP3/MP4/WAV)", file_count="single")

    # ä½¿ç”¨ Markdown ç¢ºä¿æ ¼å¼åŒ–è¼¸å‡ºæ­£ç¢ºæ¸²æŸ“ï¼ŒåŒ…æ‹¬éŒ¯èª¤è¨Šæ¯
    output = gr.Markdown(label="è¾¨è­˜çµæœ")

    btn = gr.Button("é–‹å§‹è¾¨è­˜", variant="primary")
    btn.click(
        fn=recognize_song,
        inputs=audio_input,
        outputs=output
    )

if __name__ == "__main__":
    demo.launch()
```
